## STREAMING LIVE MARKET DATA

```python
import json
import asyncio
import websockets
import datetime
import matplotlib.pyplot as plt
import matplotlib.animation as animation

url = 'wss://ws-subscriptions-clob.polymarket.com/ws/market'
last_time_pong = datetime.datetime.now()
msgs = []

# Will the eagles win the super bowl??
eagles_yes_token = "110222417228270638383974743746762302792556220380554556504458115620557107501861"
eagles_no_token = "34527047802979125804174050325432167077742263137042415948357405372259768531455"

async with websockets.connect(url) as websocket:
    await websocket.send(json.dumps({"assets_ids":[eagles_yes_token, eagles_no_token],"type":"market"}))

    while True:
        m = await websocket.recv()
        if m != "PONG":
            last_time_pong = datetime.datetime.now()
        d = json.loads(m)
        print(d)
        if last_time_pong + datetime.timedelta(seconds=10) < datetime.datetime.now():
            await websocket.send("PING")
        else:
            msgs.append(d)
```
